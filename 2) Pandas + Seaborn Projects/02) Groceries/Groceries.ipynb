## Grocery store data is given

Data of items bought from grocerey store in a time period between 2014 and 2015 is provided to us and we have to analyze this data

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd                                                                                                                                                                                                                                                                                                                                                          
```

```python
#load dataframe
groc = pd.read_csv('Groceries_dataset.csv')
groc
```

### Analyze top 10 most bought store items

```python
#Find top 10

top10 = groc['itemDescription'].value_counts().head(10).reset_index()
top10
```

```python
#Visualize the top 10 grocery store items (need categorical plot, ex barplot, but we try stripplot)

plt.figure(figsize = (8, 5), dpi = 100)

sns.set_style('darkgrid')

titleFont = {'weight':'bold', 'color':'orangered', 'size':20, 'name':'Comic Sans MS'}
normalFont = {'color':'maroon', 'size':16}

plt.hlines(y = top10.index, xmin = 0, xmax = top10['count'], color='skyblue', alpha=0.8, linewidth=3)
sns.stripplot(x = top10['count'], y = top10['itemDescription'], orient='h', size=15, color='dodgerblue')
plt.title('Top 10 grocery store items bought', fontdict = titleFont)

plt.ylabel('Grocery Items', fontdict = normalFont)
plt.xlabel('Count', fontdict = normalFont)

plt.tight_layout()
plt.savefig('Top10plot.png', dpi = 300)
plt.show()
```
