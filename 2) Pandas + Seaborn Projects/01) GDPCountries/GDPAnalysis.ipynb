# GDP Of Countries Dataset Analysis

We will be using the pandas library to analyse the given dataset of GDP of countries in the time period 2020-2025. We will be doing **Exploratory Data Analysis** (*EDA*), in which we will try to uncover various patterns/find anomalies in the dataset by analysis and visualizations. 

The given data is in terms of millions of current USD.

We shall use matplotlib for the first plot on the project, but will use seaborn as well.

```python
#import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load and see data format
gdp = pd.read_csv('2020-2025.csv')
gdp
```

#### Analysing richest and poorest countries (Considering overall GDP over the 6 years)

```python
#Creating new array sorted based on total GDP over the 5 years

gdpSorted = gdp.copy()
gdpSorted['Total'] = gdp.loc[:, '2020':'2025'].sum(axis = 1)
gdpSorted = gdpSorted.sort_values('Total', ascending = False).reset_index(drop = True)

gdpSorted
```

```python
# Getting top 10 and bottom 10 countries for anaylsis via visualization

top10 = gdpSorted.loc[:9, :]
bottom10 = gdpSorted.tail(10).sort_values('Total').reset_index(drop = True)

titleFont = {'weight': 'bold', 'size': 20, 'name' : 'Comic Sans MS'}

plt.figure(figsize = (12, 12), dpi = 100)
plt.subplot(2, 1, 1)
plt.barh(top10['Country'], top10['Total']/(10**6), color = 'green')
plt.title("GDP of the 10 richest countries over the last 6 years", fontdict = {'weight': 'bold', 'size': 20, 'name' : 'Comic Sans MS'})
plt.ylabel('Countries', size = 16)
plt.xlabel('Total GDP (Current USD in Trillions)', size = 16)
plt.grid(axis = 'x', linestyle = '--', alpha = 0.4)

plt.subplot(2, 1, 2)
plt.barh(bottom10['Country'], bottom10['Total'], color = 'red')
plt.title("GDP of the 10 poorest countries over the last 6 years", fontdict = titleFont)
plt.ylabel('Countries', size = 16)
plt.xlabel('Total GDP (Current USD in Millions)', size = 16)
#plt.xticks(rotation=30, ha='right')
plt.grid(axis = 'x', linestyle = '--', alpha = 0.4)

plt.tight_layout()
plt.savefig('GDPRichestPoorest.png', dpi = 300)
plt.show()
```

That was our analysis of the GDP of the top 10 and lowest 10 countries in terms of their GDP over the last 6 years, along with a visualization.

#### Analyzing India and its neighbouring countries

```python
#Getting the countries neighbouring India
indiaN = gdp[gdp['Country'].isin(['India', 'Myanmar', 'Bangladesh', 'Nepal', 'Bhutan', 'China', 'Pakistan', 'Maldives', 'Sri Lanka', 'Afghanistan'])].copy()
indiaN
```

```python
#Add sum and mean columns to add more interesting data to dataframe

pd.options.display.float_format = '{:.2f}'.format

indiaN['Total'] = indiaN.loc[:, '2020':'2025'].sum(axis = 1)
indiaN['Average'] = indiaN.loc[:, '2020':'2025'].mean(axis = 1)

indiaN
```

```python
#Visalizations of Total GDP of countries neighbouring india with bar plot

plt.figure(figsize = (11, 5), dpi = 100)
ax = sns.barplot(x = indiaN.Country, y = indiaN.Total, palette = 'deep', hue = indiaN.Total, legend = False)
plt.title("GDP of India and its Neighbouring Countries (log scale)", fontdict = titleFont)
plt.grid(axis = 'y', linestyle = '--', alpha = 0.6)
ax.set_yscale('log')

for container in ax.containers:
    ax.bar_label(container, fmt = '%.0f', weight = 'bold', color = 'gold')

plt.savefig('GDPindiaN.png', dpi = 300)
plt.tight_layout()
plt.show()
```

```python
#Create a melted copy of dataframe to create its lineplot

#Revisit to understand below line better (melting, pivoting, long/wide dynamics)
indiaNLong = pd.melt(indiaN, id_vars = 'Country', value_vars = ['2020', '2021', '2022', '2023', '2024', '2025'], var_name = 'Year', value_name = 'GDP')
indiaNLong.sample(10)   
```

```python
#Countries in decreasing sorted order (needed later)

countries  = indiaN.sort_values('Total', ascending = False)['Country']
countries
```

```python
#Make Lineplot with new created melted dataframe

plt.figure(figsize = (8,5), dpi = 100)
ax = sns.lineplot(data = indiaNLong, x = 'Year', y = 'GDP', hue = 'Country', hue_order = countries, palette = 'dark')
ax.set_yscale('log')
#to place legend out of the graph
ax.legend(bbox_to_anchor  = (1,1))

plt.title("GDP of countries neighbouring India over time", fontdict = titleFont)

#hue_order = countries for some reason makes legend and hues color match up, im not complaining thats awesome but no clue why that works

plt.savefig('GDPIndiaTime.png', dpi = 300)
plt.show()
```
